```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt

df=pd.read_csv("./Q1_data.csv")
df
```

```python
df=df.truncate(after=10)
df
```

```python
a=sns.barplot(x=df['Date'],y=df['Turnover'])
plt.draw()
a.set_xticklabels( a.get_xticklabels() , rotation = 90 )
print()
```

```python
a=sns.histplot(x=df['Date'],y=df['Turnover'])
plt.draw()
a.set_xticklabels( a.get_xticklabels() , rotation = 90 )
print()
```

```python
a=sns.scatterplot(x=df['Date'],y=df['Turnover'])
plt.draw()
a.set_xticklabels( a.get_xticklabels() , rotation = 90 )
print()
```
